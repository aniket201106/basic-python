# basic python

1 What is Python, and why is it popular?

  - Python is a high-level, interpreted programming language known for its simplicity, readability, and versatility. It's popular due to its large standard library, active community, and use in web development, data science, automation, AI, and more.

2 What is an interpreter in Python?

  - An interpreter executes Python code line by line. It converts high-level Python code into machine code at runtime.

3 What are pre-defined keywords in Python?

  -  These are reserved words that have special meaning in Python, like if, else, while, for, def, etc. They cannot be used as variable names.

4  Can keywords be used as variable names?

  - No, Python keywords are reserved and cannot be used as variable names.

5 What is mutability in Python?

- Mutability refers to whether an object can be changed after creation. Mutable objects (like lists) can be changed, while immutable objects (like tuples) cannot.

6  Why are lists mutable, but tuples are immutable?

  - Lists are designed to be modified after creation (add/remove items), while tuples are meant to be constant data collections, hence immutable.

7 What is the difference between “==” and “is” operators in Python?

  - == compares values, while is compares identity (memory location).

8  What are logical operators in Python?

  - and, or, and not are logical operators used to combine conditional statements.

9  What is type casting in Python?

  - Converting one data type into another. Example: int("5") converts string "5" to integer 5.

10  What is the difference between implicit and explicit type casting?

  - Implicit: Done automatically by Python (5 + 2.0 becomes 7.0)

  -  Explicit: Done manually using functions like int(), str(), etc.

11  What is the purpose of conditional statements in Python?

  - Conditional statements allow you to execute certain blocks of code based on conditions (e.g., if, elif, else).

12 How does the elif statement work?

  - elif allows checking multiple expressions for True, after an initial if statement.

13 What is the difference between for and while loops ?

  - for loop: Used when the number of iterations is known.

  - while loop: Used when the loop continues based on a condition.

14  Describe a scenario where a while loop is more suitable than a for loop?

  - When waiting for user input or monitoring a condition until it's met, a while loop is more suitable (e.g., keep asking for password until it's correct).
